## Exercise 5.10

We consider the system
\begin{align*}
x' &=  bx - xy\\
y' &= -y + xy
\end{align*}
the jacobian is 
\begin{align*}
Df(x, y)&=
\begin{bmatrix}
b - y & -x\\
y & -1 +x
\end{bmatrix}
\end{align*}
note that the coexistence fixed point is $(1, b)$. Then the linearization gives us 
\begin{align*}
Df(0, 0) &=
\begin{bmatrix}
b & 0\\
0 & -1
\end{bmatrix}, \lambda_1 = b,\, \lambda_2 = -1
\end{align*}
Therefore the origin is a saddle point with a stable direction along the eigenspace for $\lambda_1$ and an unstable direction along th eeigenspace for $\lambda_2$. Then the next point
\begin{align*}
Df(1, b) &=
\begin{bmatrix}
0 & -1\\
b & 0
\end{bmatrix}, \lambda = \pm i \sqrt{b}
\end{align*}
therefore the coexistence point is a center, i.e. a stable but not asymptotically stable node.

## Exercise 5.12

```python
#define ode system
def DF(x, y, b):
    
    return x*(x-y), y*(2*x - y)

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 300), np.linspace(-3.0, 3, 300))
u, v = np.zeros_like(X), np.zeros_like(X)
NI, NJ = X.shape

for i in range(NI):
    for j in range(NJ):
        x, y = X[i, j], Y[i, j]
        ode = DF(x, y, 2)
        u[i, j] = ode[0]
        v[i, j] = ode[-1]

x = np.linspace(-3, 3, 150)
x_nullcline = lambda x: x
y_nullcline = lambda x: (1/2.)*x
fig = plt.figure()
fig.set_dpi(300)
ax = fig.add_subplot(111)
ax.streamplot(X, Y, u, v, color='r')
ax.plot(0, 0, 'bo', markersize=4, label='equilibirum at the origin')
ax.plot(x, x_nullcline(x), 'g-', label='x nullcline')
ax.plot(x, y_nullcline(x), 'm-', label='y nullcline')
ax.legend(loc='best')
ax.axis('square')
ax.set_title('Phase Portrait for the Given System')
ax.axis([-3, 3, -3, 3])
plt.show()
```

## Exercise 5.13

Assume that $b > g> 0$ for the second order Lotka-Volterra model. The jacobian is 
\begin{align*}
Df(x, y)&=
\begin{bmatrix}
b - 2gx - y & -x\\
y & x-1
\end{bmatrix}
\end{align*}
For the origin we have 
\begin{align*}
Df(0, 0) &=
\begin{bmatrix}
b & 0 \\
0 &-1
\end{bmatrix}, \lambda_1 = b, \lambda_2 = -1
\end{align*}
for the same reason as in 5.10 the origin is a saddle point. Then for the point $(b/g, 0)$ we have 
\begin{align*}
\begin{bmatrix}
-b & -b/g\\
0 & b/g-1
\end{bmatrix}, \lambda_1 = \frac{b-g}{g}, \, \lambda_2 = -b
\end{align*}
recall that b > g meaning that $\lambda_1> 0$, so this is a saddle point as well with $\lambda_1$ the unstable direction and $\lambda_2$ the stable direction. 

Finally, 
\begin{align*}
Df(1, b-g) &=
\left[\begin{matrix}- g & -1\\b - g & 0\end{matrix}\right], \lambda =-g \pm \frac{1}{2}\sqrt{-4b + g^2 + 4g}
\end{align*}
Here we have a few cases. In the case where the square root introduces a complex value then we have that that this is a stable spiral. On the other hand if the square root is a real number with absolute value less than $2g$ then both eigen values are real and negative then we have asymptotic stability. 
Finally if the square root has absolute value larger than $2g$ then we have a saddle point.

## Exercise 5.14

For the new Lotka Voltera model we have the jacobian
\begin{align*}
Df(x, y) &=
\begin{bmatrix}
b -\frac{y}{x+y} + \frac{xy}{(x+1)^2} & \frac{-x}{x+1}\\
xy + (x+1)y & x(x+1)-2
\end{bmatrix}
\end{align*}
then at the origin we have
\begin{align*}
Df(0, 0) &=
\begin{bmatrix}
b & 0\\
0 & -2
\end{bmatrix}, \lambda_1 = b, \, \lambda_2 = -2
\end{align*}
making the origin a saddle point.

## Exercise 5.15
The final equilibrium point is $(-2, -b)$. This gives us
\begin{align*}
Df(-2, -b) &=
\begin{bmatrix}
2b & -2\\
3b & 0
\end{bmatrix}, 
\lambda = b \pm \sqrt{(b-6)b}
\end{align*}
notice that if $b=6$ then there is only one eigenvalue $b$ which we are assuming is greater than 0. In this case then the origin is unstable, if $b< 6$ then we have two complex eigenvalues with the real part positive making this an unstable spiral.
