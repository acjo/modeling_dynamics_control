```python
import numpy as np
from matplotlib import pyplot as plt
```

# Exercise 5.34

```python

x, y = np.linspace(-5, 5, 1000), np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)

Z = X + 1j * Y

Z = np.abs(1 - Z)

fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
plt.contourf(Z > 1, cmap='gray')
ax.axhline(500, color='red', alpha=0.5)
ax.axvline(500, color='red', alpha=0.5)

ax.set_aspect('equal')
plt.axis('off')
plt.show()
```

# Exericse 5.36

```python

x, y = np.linspace(-5, 5, 1000), np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)

Z = X + 1j * Y

Z = np.abs(1 + Z + Z**2/2)

fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
plt.contourf(Z < 1, cmap='gray')
ax.axhline(500, color='red', alpha=0.5)
ax.axvline(500, color='red', alpha=0.5)

ax.set_aspect('equal')
plt.axis('off')
plt.show()
```
