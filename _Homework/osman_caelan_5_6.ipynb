```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sy
from IPython.display import display
```

```python
#define ode system
def DF(x, y):
    
    return x-y-x**3, x+y-y**3

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 300), np.linspace(-3.0, 3, 300))
u, v = np.zeros_like(X), np.zeros_like(X)
NI, NJ = X.shape

for i in range(NI):
    for j in range(NJ):
        x, y = X[i, j], Y[i, j]
        ode = DF(x, y)
        u[i, j] = ode[0]
        v[i, j] = ode[-1]

x, y = np.linspace(-3, 3, 150), np.linspace(-3, 3, 150)
x_nullcline = lambda y: y**3 - y
y_nullcline = lambda x: x-x**3
fig = plt.figure()
fig.set_dpi(300)
ax = fig.add_subplot(111)
ax.streamplot(X, Y, u, v, color='r')
ax.plot(0, 0, 'bo', markersize=4, label='equilibirum at the origin')
ax.plot(x_nullcline(y),y, 'g-', label='x nullcline')
ax.plot(x, y_nullcline(x), 'm-', label='y nullcline')
ax.legend(loc='best')
ax.axis('square')
ax.set_title('Phase Portrait for the Given System')
ax.axis([-3, 3, -3, 3])
plt.show()
```
