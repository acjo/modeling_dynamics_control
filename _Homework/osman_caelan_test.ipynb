```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
def plot_canonical():
    
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    y = x.copy()
    X, Y = np.meshgrid(x, y)
    
    Z1 = 1-np.sqrt(X**2 + Y**2)
    Z2 =1+ np.sqrt(X**2 + Y**2)
    
    fig = plt.figure()
    fig.set_dpi(150)
    ax = fig.add_subplot(111, projection='3d')
    
    ax.plot_surface(X, Y, Z1, color='green')
    ax.plot_surface(X, Y, Z2, color='green')
 

    plt.show()
    
    
    return
plot_canonical()
```

```python
def plot_canonical_spherical():
    
    Î¸ = np.linspace(0, 2*np.pi, 1000)
    z = np.linspace(-2, 2, 1000)
    
    Î˜, Z = np.meshgrid(Î¸, z)
    
    X = (1+Z)**2*np.cos(Î˜)
    Y = (1+Z)**2*np.sin(Î˜)
    
    
    fig = plt.figure()
    fig.set_dpi(150)
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    
    plt.show()
    return

plot_canonical_spherical()
```
