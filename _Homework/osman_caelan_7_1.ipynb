## Part 2

```python
x, y = np.linspace(-1, 1, 1000), np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)

Z = X**2 + 1+ X 


fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
plt.contourf(Z > 0, cmap='gray')
ax.axhline(500, color='red', alpha=0.5)
ax.axvline(500, color='red', alpha=0.5)

ax.set_aspect('equal')
plt.axis('off')
plt.show()
```
