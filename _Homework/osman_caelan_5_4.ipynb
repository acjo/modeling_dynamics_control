```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

# Exercise 5.18

```python
#define Hamiltonian system
def DF(x, y):
    
    return 8*y, -2*x

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 300), np.linspace(-3.0, 3, 300))
u, v = np.zeros_like(X), np.zeros_like(X)
NI, NJ = X.shape

for i in range(NI):
    for j in range(NJ):
        x, y = X[i, j], Y[i, j]
        ode = DF(x, y)
        u[i, j] = ode[0]
        v[i, j] = ode[-1]

fig = plt.figure()
fig.set_dpi(150)
ax = fig.add_subplot(111)
ax.streamplot(X, Y, u, v, color='r')
ax.axis('square')
ax.set_title('Phase Portrait for the Hamiltonian system.')
ax.axis([-3, 3, -3, 3])
plt.show()
```
